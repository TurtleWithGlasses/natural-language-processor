### With Regex

```python
import re
```

```python
len(re.findall(pattern, book))
```

### The sentences where love is used

```python
pattern = re.compile("love")
find = re.findall(pattern, book)
```

```python
len(find)
```

```python
pattern = re.compile("[a-zA-Z]* love [a-zA-Z]*")
find = re.findall(pattern, book)
```

```python
find
```

```python
pattern = re.compile("[^.]* love [^.]*.")
find = re.findall(pattern, book)
find
```

```python
len(find)
```

```python
pattern = re.compile("[^.]* love[^a-zA-Z]+[^.]*.")
find = re.findall(pattern, book)
len(find)
```

```python
pattern = re.compile("[^.]*love[^a-zA-Z]+[^.]*.")
find = re.findall(pattern, book)
len(find)
```

```python
find
```

```python
pattern = re.compile("[A-Z]{1}[^.]*love[^a-zA-Z]+[^.]*.")
find = re.findall(pattern, book)
find
```

# What are the most used words?

```python
pattern = re.compile("[a-zA-Z]+")
find = re.findall(pattern, book.lower())
find

d = {}
for word in find:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1
```

```python
d
```

```python
import re

pattern = re.compile("[a-zA-Z]+")
find = re.findall(pattern, book.lower())

for index, word in enumerate(find, 1):
    counting = len(word)
    result = f"{index}. {word:>{counting}} : {counting}"
    print(result)
```

```python
pattern = re.compile("[a-zA-Z]+")
find = re.findall(pattern, book.lower())
find

d = {}
for word in find:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1
```

```python
d
```

```python
d_list = [(value, key) for (key, value) in d.items()]
sorted(d_list, reverse=True)
```
